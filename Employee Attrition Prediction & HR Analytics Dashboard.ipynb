```python
# Checking Missing Values in or original dataset df
df.isnull().sum().sum()
# There is no missing value we don't need to handle any
```

```python
# Seperating target column and training data into X and y
X=df.drop(columns=['Attrition'])
y=df.Attrition
```

```python
print('shape of X : ',X.shape,'shape of y : ',y.shape)
```

```python
# Lets encode the catogerical variables of X and import Standerd Scaler for feature scaling of data
from sklearn.preprocessing import LabelEncoder,StandardScaler
le=LabelEncoder()
for column in X.select_dtypes(include=['object']).columns:  # Only encode object/categorical columns
    X[column] = le.fit_transform(X[column])
```

```python
# We will scale only those values which was already a numeric column before Encoding
scaler=StandardScaler()
numerical_cols = ['Age', 'MonthlyIncome', 'DistanceFromHome', 'TotalWorkingYears', 
                  'YearsAtCompany', 'YearsWithCurrManager', 'PercentSalaryHike']
X[numerical_cols]=scaler.fit_transform(X[numerical_cols])
```

# Importing Train Test and Split

```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV,KFold
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.20,shuffle=True)
```

# Building SVM

```python
from sklearn.svm import SVC
```

```python
svc_model=SVC()
```

```python
svc_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred=svc_model.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print('Test Score ',svc_model.score(X_test,y_test))
```

```python
print('Train Score ',svc_model.score(X_train,y_train))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

def evaluate_model(model, X_test, y_test, positive_label='Yes'):
    """
    Evaluates a trained binary classification model on test data.
    
    Parameters:
    - model: trained ML model with .predict() and .predict_proba()
    - X_test: test features
    - y_test: true labels
    - positive_label: label considered as the positive class ('Yes' by default)
    """
    
    y_pred = model.predict(X_test)
    
    # If model has predict_proba
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:,1]
        roc_auc = roc_auc_score(y_test, y_prob)
    else:
        y_prob = None
        roc_auc = None

    print(" Evaluation Metrics")
    print("---------------------")
    print(f" Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
    print(f" Precision: {precision_score(y_test, y_pred, pos_label=positive_label):.4f}")
    print(f" Recall:    {recall_score(y_test, y_pred, pos_label=positive_label):.4f}")
    print(f" F1 Score:  {f1_score(y_test, y_pred, pos_label=positive_label):.4f}")
    if roc_auc is not None:
        print(f" ROC-AUC:   {roc_auc:.4f}")
    
    print("\n Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred, labels=[positive_label, 'No' if positive_label == 'Yes' else 'Yes'])
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=[positive_label, 'No' if positive_label == 'Yes' else 'Yes'],
                yticklabels=[positive_label, 'No' if positive_label == 'Yes' else 'Yes'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(' Confusion Matrix')
    plt.show()
```

```python
evaluate_model(svc_model, X_test, y_test, positive_label='Yes')
```
